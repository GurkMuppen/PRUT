```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lines = []
with open("DATALOG5.CSV","r") as f:
    lines = f.readlines()

timelist = []
xvallist = []
yvals = []
zvals = []

#xpos,ypos,zpos =0,0,0
#xvel,yvel,zvel=0,0,0
currenttime= float(lines[0].split(",")[0])/1000

for line in lines:
    print(line)
    time,xval,yval,zval=line.split(",")
    time,xval,yval,zval=float(time)/1000, float(xval),float(yval),float(zval)-10
    timelist.append(time)
    xvallist.append(xval)
    yvals.append(yval)
    zvals.append(zval)

figure, axis = plt.subplots(2, 2) 

axis[0, 0].plot(timelist, xvallist) 
axis[0, 0].set_title("X axis")

axis[1, 0].plot(timelist, yvals) 
axis[1, 0].set_title("Y axis")

axis[0, 1].plot(timelist, zvals) 
axis[0, 1].set_title("Z axis")

plt.show()
```
